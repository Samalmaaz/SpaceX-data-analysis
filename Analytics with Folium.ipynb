```python
from math import radians, sin, cos, sqrt, atan2

# Function to calculate haversine distance
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of the Earth in kilometers
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    # Compute differences
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    # Haversine formula
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

# Example proximity location: NASA Johnson Space Center (Houston, TX)
proximity_lat = 29.559684888503615
proximity_lon = -95.0830971930759

# Calculate distances and store in a new column
launch_sites_df['Distance_to_Proximity'] = launch_sites_df.apply(
    lambda row: haversine(row['Lat'], row['Long'], proximity_lat, proximity_lon), axis=1
)

# Display the updated dataframe with distances
print(launch_sites_df)
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Coordinates of CCAFS LC-40 launch site
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Coordinates of the closest coastline (based on map exploration)
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calculate the distance
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Print the distance
print(f"Distance from launch site to coastline: {distance_coastline:.2f} km")
```

```python
# Define the coordinates of the launch site (e.g., CCAFS LC-40)
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Define the coordinates of the closest coastline point
coastline_lat = 28.56367
coastline_lon = -80.57163

# Calculate the distance between the launch site and the coastline
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Initialize the map at the launch site location
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=15)

# Add the launch site marker
folium.Marker(
    [launch_site_lat, launch_site_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>Launch Site</b></div>',
    )
).add_to(site_map)

# Add the coastline marker with the calculated distance
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_coastline),
    )
)

# Add the marker to the map
site_map.add_child(distance_marker)

# Display the map
site_map
```

```python
# Define the coordinates of the launch site (e.g., CCAFS LC-40)
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Coordinates for the closest city, railway, highway, and coastline (update with precise values if available)
city_lat, city_lon = 28.3922, -80.6077  # Cocoa Beach
railway_lat, railway_lon = 28.5721, -80.5853  # Railway near the launch site
highway_lat, highway_lon = 28.5634, -80.5768  # Nearby Highway
coastline_lat, coastline_lon = 28.56367, -80.57163  # Closest coastline point

# Calculate distances from the launch site to each proximity
distance_city = calculate_distance(launch_site_lat, launch_site_lon, city_lat, city_lon)
distance_railway = calculate_distance(launch_site_lat, launch_site_lon, railway_lat, railway_lon)
distance_highway = calculate_distance(launch_site_lat, launch_site_lon, highway_lat, highway_lon)
distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Initialize the map at the launch site location
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=15)

# Add the launch site marker
folium.Marker(
    [launch_site_lat, launch_site_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>Launch Site</b></div>',
    )
).add_to(site_map)

# Add markers for the city, railway, highway, and coastline with distances
city_marker = folium.Marker(
    [city_lat, city_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_city),
    )
)
railway_marker = folium.Marker(
    [railway_lat, railway_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_railway),
    )
)
highway_marker = folium.Marker(
    [highway_lat, highway_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_highway),
    )
)
coastline_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(20, 20),
        icon_anchor=(0, 0),
        html='<div style="font-size: 12; color:#d35400;"><b>{:10.2f} KM</b></div>'.format(distance_coastline),
    )
)

# Add markers to the map
site_map.add_child(city_marker)
site_map.add_child(railway_marker)
site_map.add_child(highway_marker)
site_map.add_child(coastline_marker)

# Draw lines between the launch site and each proximity
folium.PolyLine([[launch_site_lat, launch_site_lon], [city_lat, city_lon]], color='blue', weight=2).add_to(site_map)
folium.PolyLine([[launch_site_lat, launch_site_lon], [railway_lat, railway_lon]], color='green', weight=2).add_to(site_map)
folium.PolyLine([[launch_site_lat, launch_site_lon], [highway_lat, highway_lon]], color='red', weight=2).add_to(site_map)
folium.PolyLine([[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]], color='purple', weight=2).add_to(site_map)

# Display the map
site_map
```
